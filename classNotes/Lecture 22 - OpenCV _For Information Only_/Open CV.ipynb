````python
from matplotlib import pyplot as plt
img = cv2.imread('test_image.jpg')
# opencv loads in BGR format by default, we want to show it in RGB there are 2 ways to do that
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
#plt.imshow(img[:,:,::-1])
````

````python
plt.imshow(img)
````

````python
img = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
plt.imshow(img)
````

````python
img2 = cv2.imread('puppy.jpg')
plt.imshow(img2)
````

````python
img2 = cv2.cvtColor(img2,cv2.COLOR_RGB2HSV)
plt.imshow(img2)
````

````python
img = cv2.imread('test_image.jpg',0) #gray scale
plt.imshow(img, cmap='gray')
````

````python
img.max()
````

````python
ret, imggray = cv2.threshold(img,150,255, cv2.THRESH_BINARY)
plt.imshow(imggray,cmap='gray')
````

````python
ret, imggray = cv2.threshold(img,120,255, cv2.THRESH_BINARY)
````
